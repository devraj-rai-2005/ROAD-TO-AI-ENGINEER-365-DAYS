```python
x = df[['Relative_Compactness', 'Surface_Area', 'Wall_Area', 'Roof_Area',
       'Overall_Height', 'Orientation', 'Glazing_Area',
       'Glazing_Area_Distribution']]
```

```python
df.columns
```

```python
y1 = df['Heating_Load']
y2 = df['Cooling_Load']
```

```python
import numpy as np
```

```python
y2
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = scaler.fit_transform(x)
```

```python
x_train, x_test, y1_train, y1_test, y2_train, y2_test = train_test_split(
    x, y1, y2, test_size=0.2, random_state=42
)
```

```python
y1_train.shape
```

```python
from keras.models import *
from keras.layers import *
```

```python
x = Input(shape = (8,))

d1 = Dense(128 , activation = 'relu')(x)
d2 = Dense(64 , activation = 'relu')(d1)
d3 = Dense(32 , activation = 'relu')(d2)
d4 = Dense(16 , activation = 'relu')(d3)
y1 = Dense(1 , activation = 'linear')(d4)
y2 = Dense(1 , activation = 'linear')(d4)

model = Model(inputs = x , outputs = [y1 , y2])
```

```python
model.summary()
```

```python
from keras.optimizers import *
```

```python
model.compile(
    optimizer=Adam(1e-3),
    loss=["mse", "mse"],
    metrics=[["mae"] , ["mae"]]
)
```

```python
model.fit(x_train , [y1_train, y2_train] , validation_data = (x_test , [y1_test, y2_test])  , epochs = 1500 , verbose = 1 , batch_size = 4)
```

```python
model.save('model_energy_ann.h5')
model.save('model_energy_ann.keras')
```

```python
import joblib

joblib.dump(scaler ,'scaling.pkl' )
```

```python
y_pred = model.predict(x_test)
```

```python
y1_pred = y_pred[0]
y2_pred = y_pred[1]
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
```

```python
r2_score(y1_test , y1_pred)
```

```python
r2_score(y2_test , y2_pred)
```

```python
print(mean_absolute_error(y1_test , y1_pred))
print(mean_absolute_error(y2_test , y2_pred))
```

```python
model.evaluate(x_train , [y1_train , y2_train])
```

```python
model.evaluate(x_test , [y1_test , y2_test])
```
