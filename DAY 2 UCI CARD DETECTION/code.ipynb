```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
```

```python
new_df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder

ct1 = ColumnTransformer(
    transformers = [
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['SEX', 'EDUCATION', 'MARRIAGE'])
    ],
    remainder='passthrough'
)
```

```python
new_df
```

```python
new_df['default'] = new_df['default'].map({'Y' : 1 , 'N' : 0})
```

```python
x = new_df.drop('default' , axis = 1 ) 
y = new_df['default']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x, y, test_size = 0.2 , random_state = 42)
```

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import *
```

```python
ct2 = SMOTE(random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
```

```python
ct3 = CatBoostClassifier(
    iterations=500,
    depth=8,
    learning_rate=0.05,
    loss_function='Logloss',
    eval_metric='AUC',
    random_seed=42,
    verbose=100
)
```

```python
pipe = Pipeline([
    ('p1' , ct1) , 
    # ('p2' , ct2) , 
    ('p3' , ct3) , 
])
```

```python
pipe.fit(x_train  , y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
pipe.score(x_train , y_train)
```

```python
pipe.score(x_test , y_test)
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(y_test , y_pred)
```

```python
print(classification_report(y_test , y_pred))
```

```python
print(confusion_matrix(y_test , y_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

# Probability
y_prob = pipe.predict_proba(x_test)[:,1]

# ROC AUC
print("ROC-AUC:", roc_auc_score(y_test, y_prob))

# Default threshold = 0.5
y_default = (y_prob >= 0.5).astype(int)
print("\nDefault Threshold Results:")
print(classification_report(y_test, y_default))

# Custom threshold
threshold = 0.35
y_custom = (y_prob >= threshold).astype(int)
print("\nCustom Threshold Results:")
print(classification_report(y_test, y_custom))
print(confusion_matrix(y_test, y_custom))
```
