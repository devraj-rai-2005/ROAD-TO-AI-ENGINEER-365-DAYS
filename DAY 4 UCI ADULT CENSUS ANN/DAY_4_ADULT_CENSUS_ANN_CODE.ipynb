```python
from sklearn.preprocessing import OrdinalEncoder

cat_cols = [
    'Workclass','Education','Martial_Status',
    'Occupation','Relationship','Race','Sex','Country'
]

ordinal = OrdinalEncoder(
    handle_unknown='use_encoded_value',
    unknown_value=-1
)

#  FIT ONLY ON TRAIN DATA
ordinal.fit(df[cat_cols])

#  TRANSFORM
df[cat_cols] = ordinal.transform(df[cat_cols])
test_df[cat_cols] = ordinal.transform(test_df[cat_cols])
```

```python
import joblib
joblib.dump(ordinal, "ordinal_encoder.pkl")
```

```python
from sklearn.preprocessing import StandardScaler

num_cols = [
    'Age','fnlwgt','Education_Num',
    'Capital_Gain','Capital_Loss','Hours_per_week'
]

scaler = StandardScaler()

# Fit only on train
scaler.fit(df[num_cols])

df[num_cols] = scaler.transform(df[num_cols])
test_df[num_cols] = scaler.transform(test_df[num_cols])

joblib.dump(scaler, "scaler.pkl")
```

```python
x_train = df.drop('Target' , axis =1 )
y_train = df['Target']
```

```python
x_train.shape
```

```python
x_train.shape[1]
```

```python
x_test = test_df.drop('Target' , axis =1 )
y_test = test_df['Target']
```

```python
x_train = Scaler.fit_transform(x_train)
```

```python
scaler = joblib.dump(Scaler , 'scaler.pkl')
```

```python
x_test = Scaler.transform(x_test)
```

```python
from keras.models import *
from keras.layers import *
```

```python
i = Input(shape = (14, ))

d1 = Dense(128 , activation  = 'relu')(i)
d2 = Dense(64 , activation  = 'relu')(d1)
d3 = Dense(32 , activation  = 'relu')(d2)
o1 = Dense(1 , activation = 'sigmoid')(d3)

model = Model(inputs = i , outputs = o1)
```

```python
model.summary()
```

```python
from keras.losses import *
from keras.optimizers import *
```

```python
model.compile(optimizer = Adam(1e-3), loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(x_train , y_train , validation_data = [x_test , y_test] , epochs = 250 , batch_size = 8 , verbose =1 )
```

```python
y_pred = model.predict(x_test)
```

```python
import numpy as np

y_pred = (y_pred >= 0.5).astype(int)
```

```python
from sklearn.metrics import *
```

```python
accuracy_score(y_test , y_pred)
```

```python
print(classification_report(y_test , y_pred))
```

```python
print(confusion_matrix(y_test , y_pred))
```
