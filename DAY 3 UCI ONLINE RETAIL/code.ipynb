```python
from sklearn.preprocessing import *
```

```python
ol = OrdinalEncoder(handle_unknown = 'use_encoded_value' , unknown_value = -1)
```

```python
new_df = df.copy()
```

```python
new_df['Country'] = ol.fit_transform(new_df[['Country']]).astype(int)
```

```python
new_df.head()
```

```python
new_df = new_df.sample(n = 5000 , random_state = 42).reset_index(drop = True)
```

```python
x = new_df
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1 , 11):
    
    k1 = KMeans(n_clusters = i , random_state = 42)
    
    k1.fit(x)
    
    wcss.append(k1.inertia_)
```

```python
wcss
```

```python
from kneed import KneeLocator
```

```python
KneeLocator(range(1 , 11) , wcss , direction = 'decreasing' , curve = 'convex').elbow
```

```python
n = 3 
```

```python
k_final = KMeans(n_clusters = n , random_state = 42)

y_pred= k_final.fit_predict(x)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x , y_pred)
```

```python
new_df['Cluster'] = y_pred
```

```python
from sklearn.compose import ColumnTransformer
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df = df.sample(n = 5000 , random_state = 42).reset_index(drop = True)
```

```python
df.head()
```

```python
ct1 = ColumnTransformer(
    transformers = [
        ('ct1_Country' , OrdinalEncoder(handle_unknown= 'use_encoded_value' , unknown_value = -1) , ['Country']) , 
        ('ct1_year' , OrdinalEncoder(handle_unknown= 'use_encoded_value' , unknown_value = -1) , ['year']) , 
        ] , 
    remainder = 'passthrough'
)
```

```python
ct2 = KMeans(n_clusters  = 3 , random_state = 42)
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
    ('p1' , ct1)  , 
    ('p2' , ct2)  , 
])
```

```python
X = df.copy()
```

```python
y_pred = pipe.fit_predict(X)
```

```python
X_transformed = pipe.named_steps['p1'].transform(X)
```

```python
silhouette_score(X_transformed, y_pred)
```

```python
import joblib

joblib.dump(pipe , 'uci_retail_cluster.pkl')
```
